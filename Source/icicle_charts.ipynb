```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import sys
import plotly.express as px
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
log_name='151024'
all=True
nb_pth='C:/Users/Vito/Documents/Seed_to_harvest_process_monitoring/'
if not all:
    Case = 'Idaho' #sys.argv[1]
    timelog=pd.read_hdf(nb_pth+'Event_log/log_'+Case+'_'+log_name+'_ALL_df.h5')
    usuals=np.load(nb_pth+'Source/Data/'+Case+'/masklayers/usuals.npy', allow_pickle=True)[0]
    keynames=list(usuals.keys())
    timelog=timelog[timelog['Crop'].isin(keynames)]
    title=Case
    title_s=title
else:
    #combining all cases
    logs = []
    crops=[]
    for Case in ['Idaho', 'NorthDakota', 'Colorado']:
        timelog=pd.read_hdf(nb_pth+'Event_log/log_'+Case+'_'+log_name+'_ALL_df.h5')
        usuals=np.load(nb_pth+'Source/Data/'+Case+'/masklayers/usuals.npy', allow_pickle=True)[0]
        keynames=list(usuals.keys())
        crops+=keynames
        logs.append(timelog[timelog['Crop'].isin(keynames)])
    timelog=pd.concat(logs)
    title='Idaho, NorthDakota, and Colorado'
    title_s='Idaho_NorthDakota_Colorado'
```

```python
df = pd.DataFrame(columns=['CaseID', 'Crop', 'First Snow Previous'
                                      , 'Last Snow', 'Emergence', 'Maturity'
                                      , 'Senescence', 'Dormancy', 'First Snow'])

case_ids = timelog['CaseID'].unique()
for case_id in case_ids:
    case_trace = timelog[timelog['CaseID']==case_id]
    if len(case_trace)==7:
        df = df.append({'CaseID': case_id
                        , 'Crop': case_trace['Crop'].values[0]
                        , 'First Snow Previous': case_trace.iloc[0]['Activity'].split(' ')[0]
                        , 'Last Snow': case_trace.iloc[1]['Activity'].split(' ')[0]
                        , 'Emergence': case_trace.iloc[2]['Activity'].split(' ')[0]
                        , 'Maturity': ' '#case_trace.iloc[3]['Activity'].split(' ')[0]
                        , 'Senescence': ' '#case_trace.iloc[4]['Activity'].split(' ')[0]
                        , 'Dormancy': case_trace.iloc[5]['Activity'].split(' ')[0]
                        , 'First Snow': case_trace.iloc[6]['Activity'].split(' ')[0]}
                        , ignore_index=True)
```

```python
def get_color(value):
    if value == 'Late':
        return 'red'
    elif value == 'Usual':
        return 'green'
    elif value == 'Early':
        return 'blue'
    else:
        return 'grey'

fig = px.icicle(df, path=[px.Constant('All'), 'First Snow Previous', 'Last Snow', 'Emergence', 'Maturity', 'Senescence', 'Dormancy', 'First Snow'], labels={'id': 'Crop'})

fig.update_traces(root_color="lightgrey", marker=dict(colors=[get_color(value) for value in fig.data[0].labels]))

fig.update_layout(
    width=1080,
    height=600,
    title=title,
    margin=dict(t=50, l=25, r=25, b=25),
    annotations=[
        dict(x=0.127, y=1.05, text="First Snow Previous", showarrow=False, xref="paper", yref="paper", font=dict(size=12, color="black")),
        dict(x=0.283, y=1.05, text="Last Snow", showarrow=False, xref="paper", yref="paper", font=dict(size=12, color="black")),
        dict(x=0.435, y=1.05, text="Emergence", showarrow=False, xref="paper", yref="paper", font=dict(size=12, color="black")),
        dict(x=0.565, y=1.05, text="Maturity", showarrow=False, xref="paper", yref="paper", font=dict(size=12, color="black")),
        dict(x=0.725, y=1.05, text="Senescence", showarrow=False, xref="paper", yref="paper", font=dict(size=12, color="black")),
        dict(x=0.845, y=1.05, text="Dormancy", showarrow=False, xref="paper", yref="paper", font=dict(size=12, color="black")),
        dict(x=0.973, y=1.05, text="First Snow", showarrow=False, xref="paper", yref="paper", font=dict(size=12, color="black"))
    ]
)

fig.show()
```
